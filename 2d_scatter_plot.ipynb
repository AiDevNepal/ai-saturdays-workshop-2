# Importing modulse

```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

# Creating x co-ordinates

```python
x_min = -15
x_max = 15
x = np.arange(x_min, x_max)
# x
```

# Creating y co-ordinates

```python
y_min = -15
y_max = 15
y = np.arange(y_min, y_max)
# y
```

# Meshing the x and y co-ordinates

```python
X, Y = np.meshgrid(x, y)

print("Shape of X: ", X.shape)
print("Shape of Y: ", Y.shape)

print()
# print("X: \n", X)
print()
# print("Y: \n", Y)
```

# Flattening the X and Y co-ordinates

```python
X = X.flatten()
Y = Y.flatten()

# print("X: \n", X)
# print("Y: \n", Y)
```

```python
plt.figure(figsize=(11, 11))
plt.scatter(X, Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Cartesian 2D plane")
```

# Creating circle labels for Scatter Plot

```python
def create_circle_label(x, y, r):
    
    output = []
    
    for i, j in zip(x, y):
        
        if i**2+j**2 <= r**2:
            output.append(1)
        else:
            output.append(0)
            
    return output    
```

```python
circle_label = create_circle_label(x=X, y=Y, r=8)
# circle_label
```

```python
plt.figure(figsize=(11, 11))
plt.scatter(X, Y, c=circle_label)
plt.xlabel("x coordinates")
plt.ylabel("y coordinates")
plt.title("Circular area labels")
```
